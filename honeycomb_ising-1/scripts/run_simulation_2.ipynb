```julia
# デバッグ用チェック（ノートブックに貼る）
using Random
# 既に include していれば不要
# include(joinpath(@__DIR__, "run_simulation.jl"))
using .HoneycombIsing

L = 20
rng = MersenneTwister(1)

# 1) 近傍リストの分布（内部は3になっているか）
nb = HoneycombIsing.Lattice.build_neighbor_list(L)
println("近傍数の分布: ", sort(unique(map(length, nb))))

# 2) 境界が +1 固定か（初期化直後）
s = HoneycombIsing.init_spins_fixed(L; ordered=false, rng=rng)
println("端行の例 s[1,:] = ", s[1,:])
println("端列の例 s[:,1] = ", s[:,1])

# 3) 任意サイトの近傍確認と ΔE（β=高値で低温相）を表示
i,j = 10,10
println("site ($i,$j) neighbours = ", nb[(i-1)*L + j])
println("spin at site = ", s[i,j])
println("ΔE(sample) = ", HoneycombIsing.local_energy_change_fixed(s, i, j; J=1.0, h=0.0, neighbor_list=nb))

# 4) 受理率の簡易測定（同じ rng を使う）
accepted = 0; attempts = 0
β = 5.0  # 低温相の例
for _ in 1:5000
    ii = rand(rng, 2:(L-1)); jj = rand(rng, 2:(L-1))
    attempts += 1
    ΔE = HoneycombIsing.local_energy_change_fixed(s, ii, jj; J=1.0, h=0.0, neighbor_list=nb)
    if ΔE <= 0 || rand(rng) < exp(-β * ΔE)
        accepted += 1
    end
end
println("accepted/attempts = ", accepted, "/", attempts, "  => rate=", accepted/attempts)

# 5) 熱化で磁化が増えるか（ordered=false から）
s2 = HoneycombIsing.init_spins_fixed(L; ordered=false, rng=MersenneTwister(2))
println("初期内部平均磁化 = ", HoneycombIsing.mag_per_spin(s2))
for t in 1:200
    HoneycombIsing.metropolis_sweep_fixed!(s2, β; J=1.0, h=0.0, rng=rng, neighbor_list=nb)
end
println("熱化後内部平均磁化 = ", HoneycombIsing.mag_per_spin(s2))

# 6) ordered=true で始めても崩れるか確認
s_ord = HoneycombIsing.init_spins_fixed(L; ordered=true, rng=MersenneTwister(3))
println("ordered=true 初期磁化 = ", HoneycombIsing.mag_per_spin(s_ord))
# 一度だけ少し動かす
HoneycombIsing.metropolis_sweep_fixed!(s_ord, β; J=1.0, h=0.0, rng=rng, neighbor_list=nb)
println("少し動かした後 ordered磁化 = ", HoneycombIsing.mag_per_spin(s_ord))
```

```julia
# ノートブックのセルに貼る（デバッグ観察）
using Random, Plots
rng = MersenneTwister(1)

L = 20
nb = HoneycombIsing.Lattice.build_neighbor_list(L)

# 初期化（ランダム内部、境界 +1）
s = HoneycombIsing.init_spins_fixed(L; ordered=false, rng=rng)

β = 10.0   # 低温の例（T=0.1）
nsweeps = 20000
record_interval = 100
mags = Float64[]

for sweep in 1:nsweeps
    HoneycombIsing.metropolis_sweep_fixed!(s, β; J=1.0, h=0.0, rng=rng, neighbor_list=nb)
    if sweep % record_interval == 0
        push!(mags, HoneycombIsing.mag_per_spin(s))
    end
end

println("最終内部平均磁化 = ", HoneycombIsing.mag_per_spin(s))
plot(collect(record_interval:record_interval:nsweeps), mags, xlabel="sweep", ylabel="m", title="magnetization vs sweep")
heatmap(Float64.(s), c=:thermal, title="final spins")
```

```julia
# ノートブックのセルに貼る（デバッグ用）
using Random
rng = MersenneTwister(1)

L = 10
nb = HoneycombIsing.Lattice.build_neighbor_list(L)

# 代表サイトを選ぶ（内部）
i,j = 5,5
idx = (i-1)*L + j

# スピン配置を作る：境界 +1、内部を -1 にして極端な場合を作る
s = ones(Int8, L, L)
for ii in 2:(L-1), jj in 2:(L-1)
    s[ii,jj] = Int8(-1)
end

println("site = ($i,$j), idx = ", idx)
println("spin at site = ", s[i,j])
println("neighbours indices = ", nb[idx])

# 近傍のスピン値と合計
nb_vals = [ s[div(n-1,L)+1, (n-1) % L + 1] for n in nb[idx] ]
println("neighbour spins = ", nb_vals)
println("nb_sum = ", sum(nb_vals))

# ΔE を計算（local_energy_change_fixed を使う）
ΔE = HoneycombIsing.local_energy_change_fixed(s, i, j; J=1.0, h=0.0, neighbor_list=nb)
println("local_energy_change_fixed -> ΔE = ", ΔE)

# メトロポリス受理判定（β 大きめ）
β = 10.0
println("exp(-β*ΔE) = ", exp(-β*ΔE))
accepted = (ΔE <= 0) || (rand(rng) < exp(-β*ΔE))
println("accept判定 (should flip if true) = ", accepted)

# 手動で flip した結果確認
s2 = copy(s)
if accepted
    s2[i,j] = -s2[i,j]
end
println("spin before -> after = ", s[i,j], " -> ", s2[i,j])
```
