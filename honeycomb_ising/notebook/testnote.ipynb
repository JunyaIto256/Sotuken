```julia
import Mypackage
using Mypackage
using Plots
using LaTeXStrings
```

```julia
# --- メイン: 温度スキャン ---

L = 32
J = 1.0
h = 0.0
n_eq = 1000
n_mc = 5000
measure_interval = 20

# 温度範囲: 0.5〜5.0
Tlist = 0.5:0.25:5.0
mags_T = Float64[]
mmags_T = Float64[]
println("=== 2D Ising 固定端境界 ===")
println("L=$L, J=$J, T ∈ [$(minimum(Tlist)), $(maximum(Tlist))]")

for T in Tlist
    mag, mmag = Mypackage.run_ising_fixed(L, T; J=J, h=h, n_eq=n_eq, n_mc=n_mc, measure_interval=measure_interval)
    push!(mags_T, mag)
    push!(mmags_T, mmag)
    println("T=$(round(T, digits=2)) → ⟨|m|⟩ = $(round(mag, digits=4)), ⟨|mm|⟩ = $(round(mmag, digits=4))")
end
```

```julia
# プロット: 磁化 vs 温度
plot(Tlist, mags_T, lw=2, marker=:circle, xlabel="Temperature T (J=1)", ylabel="⟨|m|⟩", title="Spontaneous Magnetization (Fixed Boundary)", legend=false)
```

```julia
# プロット: スピン相関 vs 温度
plot(Tlist, mmags_T, lw=2, marker=:circle, xlabel="Temperature T (J=1)", ylabel="⟨|mm|⟩", title="Spin Correlation (Fixed Boundary)", legend=false)
```
